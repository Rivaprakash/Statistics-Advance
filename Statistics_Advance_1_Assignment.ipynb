* Discrete: Takes specific values (e.g., number of heads in a coin toss).

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def bernoulli_pdf(p, x):
    """Calculate the probability distribution function (PDF) of a Bernoulli distribution."""
    return stats.bernoulli.pmf(x, p)

p = 0.5  # Probability of success
x = [0, 1]  # Possible values in a Bernoulli distribution
pdf_values = bernoulli_pdf(p, x)

# Plot Bernoulli PDF
plt.bar(x, pdf_values, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution PDF')
plt.xticks([0, 1])
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Central Limit Theorem Simulation
def clt_simulation(sample_size=30, num_samples=1000):
    original_dist = np.random.exponential(scale=2, size=10000)
    sample_means = [np.mean(np.random.choice(original_dist, sample_size)) for _ in range(num_samples)]

    sns.histplot(sample_means, bins=30, kde=True, color='purple')
    plt.title('CLT Simulation (Exponential Distribution)')
    plt.show()

clt_simulation()
```

```python
#11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Verify Central Limit Theorem with Normal Distribution
def verify_clt(sample_size=30, num_samples=1000):
    population = np.random.normal(loc=5, scale=2, size=10000)
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

    sns.histplot(sample_means, bins=30, kde=True, color='blue')
    plt.title('CLT Verification (Normal Distribution)')
    plt.show()

verify_clt()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

# Z-test function
def z_test(sample, pop_mean, pop_std):
    z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    return z, 2 * (1 - stats.norm.cdf(abs(z)))

# Sample data & Z-test
np.random.seed(42)
sample = np.random.normal(50, 10, 30)
z, p = z_test(sample, 50, 10)

print(f"Z-score: {z:.2f}, P-value: {p:.4f}")
print("Reject null" if p < 0.05 else "Fail to reject null")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Confidence Interval Function
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2, df=len(sample)-1) * sem
    return mean - margin, mean + margin

# Generate sample data
np.random.seed(42)
sample = np.random.normal(50, 10, 30)

# Compute Confidence Interval
ci_lower, ci_upper = confidence_interval(sample)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies within this interval.")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

# Generate normal distribution data
np.random.seed(42)
data = np.random.normal(50, 10, 30)

# Compute confidence interval
def confidence_interval(data, confidence=0.95):
    mean, sem = np.mean(data), stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2, df=len(data)-1) * sem
    return mean - margin, mean + margin

ci_lower, ci_upper = confidence_interval(data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Interpretation: The true mean likely falls within this range with 95% confidence.")
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate dataset
np.random.seed(42)
data = np.random.normal(50, 15, 1000)

# Apply Z-score normalization
z_scores = (data - np.mean(data)) / np.std(data)

# Visualization
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, alpha=0.5, label='Original Data', color='blue')
plt.hist(z_scores, bins=30, alpha=0.5, label='Z-score Normalized Data', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Original and Z-score Normalized Data')
plt.legend()
plt.show()

# Print summary statistics
print(f"Original Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data):.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Standard Deviation: {np.std(z_scores):.2f}")
```
